```python
import matplotlib.pyplot as plt
import geopandas as gpd
import mplleaflet
```

```python
points_fp = "Municipal_Boundaries_of_Italy_2019-shp/Com01012019_WGS84.shp"
points = gpd.read_file(points_fp)
```

```python
import folium
```

```python
m = folium.Map(location=[40.66, 14.42], zoom_start=6, control_scale=True, prefer_canvas=True)
#points_gjson.add_to(m)
#m
```

```python
#1754 vico, 5982 troia, 688 san casciano, 6557 isernia, 7474 marineo, 5298 stigliano, 6960 fubine, 7331 Mussomeli, 7482 Ciminna, 6174 Campagna, 3366 castelluccio valmaggiore, 1001 mariglianella, 987 somma vesuviana, 2561 moliterno
my_comunes = points.iloc[[1754, 5982, 688, 6557, 7474, 5298, 6960, 7331, 7481, 6174, 3366, 1001, 987, 2561]]

ancestors = ['Gaetano (Thomas) Imperato, born 1896', 'Joseph Brasile, born 1895', 'Olimpio Capitini, born 1864', 'Valentina Romano, born 1871', 'Rosina Realmuto, born 1891', 'Domenico DeCrescenzo, born 1888', 'Fiorentino Barbero, born 1896', 'Salvatore Costanzo, born 1887', 'Francesco Adamo, born 1877', 'Michele Magliano, born 1875', 'Maria Donata Marchese, born 1870', 'Vito De Crescenzo, born 1856', 'Francesca Ferraro, born 1819', 'Isabella Orlando, born 1802']
line = ['Imperato', 'Brasile', 'Imperato', 'Imperato', 'Imperato', 'Brasile', 'Brasile', 'Brasile', 'Imperato', 'Brasile', 'Brasile', 'Brasile', 'Brasile', 'Brasile']
my_comunes['Ancestors'] = ancestors
my_comunes['Line'] = line

style_function = lambda x: {'color': '#ff6969' if x['properties']['Line']=='Brasile' else '#3388FF'}
gjson = folium.features.GeoJson(my_comunes, name="comunes", style_function = style_function).add_to(m)

folium.features.GeoJsonPopup(fields=['COMUNE','Ancestors'],labels=False).add_to(gjson)
m
```
